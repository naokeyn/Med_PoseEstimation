```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/app/data/Boxes_FrontRight_4th.csv")
df["id"] = df["id"].astype(int)
df["class"] = df["class"].astype(int)
df.head()
```

```python
p1 = df[df["id"] == 1]
p5 = df[df["id"] == 8]
p11 = df[df["id"] == 11]
p13 = df[df["id"] == 13]


fig = plt.figure(figsize=(12, 12))
fig.subplots_adjust(hspace=0.3)

for i, pos in enumerate(["x1", "y1"]):
    ax = fig.add_subplot(2, 1, i+1)
    ax.plot(p1["frames"], p1[pos], label="id: 1")
    ax.plot(p5["frames"], p5[pos], label="id: 5")
    ax.plot(p11["frames"], p11[pos], label="id: 11")
    ax.plot(p13["frames"], p13[pos], label="id: 13")
    ax.legend()
    ax.set_xlabel("Frames")
    ax.set_ylabel(pos.upper())
    ax.grid()


plt.show()
```
